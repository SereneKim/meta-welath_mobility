````python
import re

dic = {}

for i, row in data[data['Q2'] == 'No'].iterrows(): # Change it to 'No' or 'Yes' to update the abstracts
    abs_text = row['abs'].strip()
    abs_text = re.sub(r'[^\x00-\x7F]+', '', abs_text)  # Remove non-ASCII characters (like Korean)
    abs_text = re.sub(r'\bp\.\s*\d+\b', '', abs_text)  # Remove page numbers
    abs_text = re.sub(r'[A-Z][a-z]+(?:\s+(?:and|&)\s+[A-Z][a-z]+)*(?:,\s+[A-Z][a-z]+)*\s*\(\d{4}\)', '', abs_text)  # Remove parenthesis citation
    abs_text = re.sub(r'[A-Z][a-z]+(?:\s+(?:and|&)\s+[A-Z][a-z]+)*(?:,\s+[A-Z][a-z]+)*(?:,\s*\d{4})', '', abs_text)  # Remove text citation
    
    abs_text = re.sub(r'<[^>]+>', '', abs_text)  # Remove HTML tags
    abs_text = re.sub(r'\b(?:ABSTRACT|Abstract)\b', '', abs_text) # Remove the term 'ABSTRACT' or 'Abstract' (case-sensitive, preserving other variations)
    abs_text = re.sub(r'https?://\S+', '', abs_text)
    abs_text = re.sub(r'www\.\S+', '', abs_text)
    abs_text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b', '', abs_text)
    abs_text = re.sub('intergenerational elasticity', 'IGE', abs_text, flags=re.IGNORECASE)
    # Normalize white spaces
    abs_text = re.sub(r'(?<!\w)[.,]+(?!\w)', '', abs_text)  # Remove isolated periods and commas
    abs_text = re.sub(r'\s+', ' ', abs_text).strip()
    
    
    dic[str(i)] = abs_text
````

### Get representative categories

````python
# ‚õîÔ∏è Check the file name before running this cell.
with open(f'{path}/data_abstracts/abstracts_Q2-No.json', 'r') as f:
    json_abs = json.load(f)
````

````python
import logging

start_time = time.time()

completion = openai_client.beta.chat.completions.parse(
        model= gpt_model,
        messages=[
            { 
            "role": "assistant",
            "content": "You are a research assistant in the Social Sciences domain. I, the researcher, have a list of abstracts of papers that studied inter/multi generational wealth/income/earning mobility. I want you to categorize the measures that the papers used based on the abstracts."
            
            },
            {
                "role": "user", 
                "content":f"""                        
                        ```Abstracts
                        {json_abs}
                        ```
                        First, find the representative categories of the measures that the papers used based on the abstracts. Here, 'measures' mean an estimate from a model or an equation that quantifies the inter/multi generational wealth/income/earning mobility.
                        Then, return the categories and lists of id's that belong to each in a JSON format.
                        For the id's that do not belong to any category, return them in a separate category named 'Others'.
                        """
            }
        ],
        response_format={"type": "json_object"},
    )
results = completion.choices[0].message.content
    
logging.info('‚úÖ Done! The process time is:', time.time()-start_time)
````

## Then ask to categorize again -- this time, one by one
We approach it this way to be robust

````python
from pydantic import BaseModel

class GetCat(BaseModel):
    category: int
    measure: str
    
````

#### Individual-level categorization

````python
cat = {}
# re-try for the papers that were not categorized yet
# with open(f'{path}/data_abstracts/categorizations_step2.json', 'r') as f:
    # cat = json.load(f)

for i, row in data[(data['Q2'] == 'No')].iterrows(): # Update it to 'No' or 'Yes'
    print(f'Processing {row.id}')
    start_time = time.time()
    if str(i) in json_abs.keys(): # and str(i) not in cat.keys():
        completion = openai_client.beta.chat.completions.parse(
                model= gpt_model,
                messages=[
                    { 
                    "role": "assistant",
                    "content": "You are a research assistant in the Social Sciences domain. I, the researcher, have an abstract of a paper that studied inter/multi generational wealth/income/earning mobility. I want you to categorize the measure(s) that the paper used based on the abstract."
                    
                    },
                    {
                        "role": "user", 
                        "content":f"""                        
                                ```Abstracts
                                {json_abs[str(i)]}
                                ```
                                First, recognize what measure(s) the papers used based on the abstracts. Here, 'measures' mean an estimate from a model or an equation that quantifies the inter/multi generational wealth/income/earning mobility.
                                Does it belong to any of the following categories?
                                1. Regression‚Äêbased Measures 
                                2. Rank‚Äêbased Measures
                                3. Transition Matrix / Probability Measures
                                4. Absolute Mobility Measures
                                5. Multigenerational Measures
                                6. Decomposition / Structural Approaches
                                7. Non‚Äêparametric Approaches
                                8. Others
                                
                                Return the category that the measure(s) belong to as a number, and the exact name of the measure(s) if possible. For example, if it belongs to '1', return also the name of the measure(s) such as 'Intergenerational Elasticity (IGE)'. 
                                Don't repeat the name of category here like 'Regression‚Äêbased Measures - IGE'.
                                If the measure(s) do not belong to any category, return '8' and the exact name of the measure.
                                """
                    }
                ],
                response_format=GetCat,
            )
        results = completion.choices[0].message.content
        cat[i] = results
    else:
        print(f'{row.id} is already processed.')       
        
logging.info('‚úÖ Done! The process time is:', time.time()-start_time)
````

````python
import logging

start_time = time.time()

with open(f'{path}/data_abstracts/cat_measures_Q2-No.json', 'r') as f:
    cat = json.load(f)
    
completion = openai_client.beta.chat.completions.parse(
        model= gpt_model, #o3-mini
        messages=[
            { 
            "role": "assistant",
            "content": "You are a research assistant in the Social Sciences domain. I, the researcher, have a description of the measure(s) that a paper used to study inter/multi-generational wealth/income/earning mobility. I want you to categorize the measures further."
            
            },
            {
                "role": "user", 
                "content":f"""
                        The following is the description of the measure(s) used in the paper.                        
                        ```json
                        {cat}
                        ```
                        
                        This was the categorization used for the measures previously.
                        1. Regression‚Äêbased Measures 
                        2. Rank‚Äêbased Measures
                        3. Transition Matrix / Probability Measures
                        4. Absolute Mobility Measures
                        5. Multigenerational Measures
                        6. Decomposition / Structural Approaches
                        7. Non‚Äêparametric Approaches
                        8. Others
                        
                        Now, based on the description of the measure(s), categorize them further.
                        'Unspecified': If the description is not clear enough to categorize, return '0'. Otherwise, '1'.
                        'Relevance': If the measure(s) are not used to study inter/multi-generational wealth/income/earning mobility, return '0'. Otherwise, '1'.
                        'Category_2': If '1' in Q1 and there is more than one type of measure mentioned, return the category that the measure(s) belong to as a number. Otherwise, return empty.
                        'Measure_1': If '1' in Q1, give me the keyword that best describes the measure(s). If there are multiple measures, return the most relevant one.
                        'Measure_2': If '1' in Q1 and there is more than one type of measure mentioned, return the second most relevant keyword. Otherwise, return empty.
                        
                        Return the results in a JSON format for each id. Each id should have ['Unspecified', 'Relevance', 'Category_2', 'Measure_1', 'Measure_2'].
                        Make sure to be consistent with the keywords. For example, if you gave 'Intergenerational Elasticity' for id '1', give the same value to id '5' whose measure is described as 'IGE' or 'IGE in income'.
            """
            }
        ],
        response_format={"type": "json_object"},
    )
results = completion.choices[0].message.content
        
logging.info('‚úÖ Done! The process time is:', time.time()-start_time)
````

````python
part1 = pd.read_csv(f'{path}/data_abstracts/Step3.csv')
part2 = pd.read_csv(f'{path}/data_abstracts/Step3_rest.csv')
cat_df = pd.read_csv(f'{path}/data_abstracts/cat_measures.csv')
len(part1), len(part2), len(cat_df)
````

````python
parts = pd.concat([part1, part2])
parts.duplicated('id').sum()
````

````python
cat_df = cat_df.drop(columns=['measure_1', 'measure_2'])
merged = cat_df.merge(parts, on='id', how='left')
````

````python
merged
````

````python
# Remember: in Part1, category_2 had more than one values (e.g., '1,2,3' or '1,2').
# I'll split them and keep the first value only.

for row in merged[(merged['category_2'].notnull()) & merged['category_2'].apply(lambda x: isinstance(x, str))].itertuples(index=False):
    if ',' in row.category_2:  # Use dot notation instead of `row['category_2']`
        split_values = row.category_2.split(',')
        split_values = [int(val) for val in split_values]
        
        first = split_values[0]
        second = split_values[1] if len(split_values) > 1 else np.nan
        third = split_values[2] if len(split_values) > 2 else np.nan
        if first == int(row.category):
            cat2 = second
            cat3 = third
        elif second == int(row.category):  
            cat2 = first
            cat3 = third
        elif third == int(row.category):
            cat2 = first
            cat3 = second
        merged.loc[merged['id'] == row.id, 'category_2'] = cat2
        merged.loc[merged['id'] == row.id, 'category_3'] = cat3
````

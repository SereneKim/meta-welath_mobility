# Neo4j

```python
from neo4j import GraphDatabase
import pandas as pd
from tqdm import tqdm
import re

# Database connection details
URI = "bolt://localhost:7687"
USERNAME = "neo4j"
PASSWORD = "your_password"

class Neo4jDatabase:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def create_indexes(self):
        """Create indexes and constraints for better performance."""
        constraints = [
            "CREATE CONSTRAINT IF NOT EXISTS FOR (p:Paper) REQUIRE p.id IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (m:Measure) REQUIRE m.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (y:Year) REQUIRE y.year IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (c:Category) REQUIRE c.name IS UNIQUE;",
        ]

        with self.driver.session() as session:
            for constraint in constraints:
                try:
                    session.run(constraint)
                    print(f"Constraint created: {constraint}")
                except Exception as e:
                    print(f"Error creating constraint: {constraint}\n{e}")

    def create_paper(self, tx, paper):
        """Insert a single paper into the database with a dynamic label."""
        # label = paper["type"]  

        query = f"""
        MERGE (p:Paper {{id: $id}})
        SET p.title = $title, p.year = $year, p.doi = $doi, 
            p.is_oa = $is_oa, p.oa_status = $oa_status, 
            p.ref_count = $ref_count, p.cited_by_count = $cited_by_count,
            p.type = $type, p.type_by4o = $type_by4o,
            p.measure_1 = $measure_1, p.measure_2 = $measure_2,
            p.category_1 = $category_1, p.category_2 = $category_2,
            p.measure = $measure
        """
        tx.run(query, **paper)

    def create_year(self, tx, year):
        """Insert a single year node."""
        query = "MERGE (y:Year {year: $year})"
        tx.run(query, year=year)

    def create_measure(self, tx, measure):
        """Insert a single measure node."""
        query = "MERGE (m:Measure {name: $name})"
        tx.run(query, name=measure)

    def create_category(self, tx, category):
        """Insert a single category node."""
        query = "MERGE (c:Category {name: $name})"
        tx.run(query, name=category)

    def create_relationships(self, tx, paper_rel):
        """Create multiple relationships for a single paper."""
        query = """
        MERGE (p:Paper {id: $id})
        
        FOREACH (y IN $year |
        MERGE (yearNode:Year {year: y})
        MERGE (p)-[:PUBLISHED_IN]->(yearNode)
        )

        FOREACH (cited_id IN $cited_papers |
            MERGE (cited_paper:Paper {id: cited_id})
            MERGE (p)-[:CITES]->(cited_paper)
        )

        FOREACH (measure IN $measures |
            MERGE (m:Measure {name: measure})
            MERGE (p)-[:USES]->(m)
        )

        FOREACH (category IN $categories |
            MERGE (c:Category {name: category})
            MERGE (p)-[:IS_CATEGORIZED_AS]->(c)
        )
        """
        tx.run(query, **paper_rel)


# Helper function to extract OpenAlex IDs from citations
def extract_openalex_ids(cited_by_str):
    if pd.isna(cited_by_str) or cited_by_str.strip() == "":
        return []
    return [url.strip() for url in cited_by_str.split(",") if url.strip()]


# Initialize Neo4j connection
neo4j_db = Neo4jDatabase(URI, USERNAME, PASSWORD)

# Create indexes before inserting data
neo4j_db.create_indexes()

# Insert Papers FIRST
print("Inserting Papers...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Papers"):
        paper = {
            "id": row["id"],
            "title": row.get("title", ""),
            "year": row.get("year", 0),
            "doi": row.get("doi", ""),
            "is_oa": row.get("is_oa", False),
            "oa_status": row.get("oa_status", ""),
            "ref_count": row.get("ref_count", 0),
            "cited_by_count": row.get("cited_by_count", 0),
            "type_by4o": row["sort_gpt4o_1"],  # Ensure safe label
            "type": row.get("type", ""),
            "category_1": row.get("category_1", ""),
            "category_2": row.get("category_2", ""),
            "measure_1": row.get("measure_1", ""),
            "measure_2": row.get("measure_2", ""),
        }
        session.write_transaction(neo4j_db.create_paper, paper)

# Insert Year Nodes
print("Inserting Years...")
with neo4j_db.driver.session() as session:
    for year in tqdm(df["year"].dropna().unique(), desc="Years"):
        if year > 0:  # Remove invalid years
            session.write_transaction(neo4j_db.create_year, year)

# Insert Measure Nodes
print("Inserting Measures...")
with neo4j_db.driver.session() as session:
    for measure in tqdm(set(df["measure_1"].dropna().unique()) | set(df["measure_2"].dropna().unique()), desc="Measures"):
        if measure.strip():
            session.write_transaction(neo4j_db.create_measure, measure)


# Insert Category Nodes (handling NaN values properly)
print("Inserting Categories...")
with neo4j_db.driver.session() as session:
    categories = set(df["category_1"].dropna().astype(str).unique()) | set(df["category_2"].dropna().astype(str).unique())

    for category in tqdm(categories, desc="Categories"):
        if category and category.strip().lower() not in {"nan", "none"}:  # Ensure it's not NaN
            session.write_transaction(neo4j_db.create_category, category)

# Insert Relationships AFTER ensuring all nodes exist
print("Inserting Relationships...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Relationships"):
        paper_rel = {
            "id": row["id"],
            "year": row.get("year", 0),
            "cited_papers": extract_openalex_ids(row.get("cited_by", "")),
            "measures": [m for m in [row.get("measure_1", ""), row.get("measure_2", "")] if pd.notna(m) and m.strip()],
            "categories": [c for c in [row.get("category_1", ""), row.get("category_2", "")] if pd.notna(c) and c.strip().lower() not in {"nan", "none"}]
        }

        session.write_transaction(neo4j_db.create_relationships, paper_rel)

# Close the database connection
neo4j_db.close()

print("Data successfully inserted into Neo4j!")
```

```python
embedding_matrix = np.array(list(measure_embeddings.values()))
embedding_matrix.shape
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import umap
from sklearn.preprocessing import StandardScaler

# Standardize the embedding matrix (Normally, they should all have the same dimension?)
scaler = StandardScaler()
embedding_matrix_scaled = scaler.fit_transform(embedding_matrix)

# Reduce to 10 dimensions for clustering
umap_reducer = umap.UMAP(n_components=8, random_state=42)
embedding_matrix_reduced = umap_reducer.fit_transform(embedding_matrix_scaled)

# Apply KMeans after UMAP
kmeans = KMeans(n_clusters=8, random_state=42)
cluster_labels = kmeans.fit_predict(embedding_matrix_reduced)
```

```python
# For visualization

# Reduce to 2D for visualization
umap_2d = umap.UMAP(n_components=2, random_state=42)
embedding_2d = umap_2d.fit_transform(embedding_matrix_scaled)

# Reduce to 3D for visualization (optional)
umap_3d = umap.UMAP(n_components=3, random_state=42)
embedding_3d = umap_3d.fit_transform(embedding_matrix_scaled)
```

```python
# Store in DataFrame
df_plot = pd.DataFrame({
    "x": embedding_2d[:, 0],
    "y": embedding_2d[:, 1],
    "z": embedding_3d[:, 2] if embedding_3d.shape[1] == 3 else np.zeros(len(embedding_2d)),
    "cluster": cluster_labels,  # From K-Means
    "measure": list(measure_embeddings.keys())  # Measure names
})
```

```python
import plotly.express as px

fig = px.scatter(
    df_plot, 
    x="x", y="y", 
    color=df_plot["cluster"].astype(str),  # Convert to string for categorical coloring
    hover_data=["measure"],  # Show measure names on hover
    title="UMAP + K-Means Clustering of Measures"
)
fig.show()
```

```python
fig = px.scatter_3d(
    df_plot, 
    x="x", y="y", z="z", 
    color=df_plot["cluster"].astype(str),
    hover_data=["measure"],
    title="3D Visualization of Measure Clusters"
)
fig.show()
```

```python
from neo4j import GraphDatabase
import pandas as pd
from tqdm import tqdm

# Database connection details
URI = "bolt://localhost:7687"
USERNAME = "neo4j"
PASSWORD = "your_password"

class Neo4jDatabase:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def create_indexes(self):
        """Create indexes and constraints for better performance."""
        constraints = [
            "CREATE CONSTRAINT IF NOT EXISTS FOR (p:Paper) REQUIRE p.id IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (m:Measure) REQUIRE m.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (y:Year) REQUIRE y.year IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (c:Category) REQUIRE c.name IS UNIQUE;",
        ]

        with self.driver.session() as session:
            for constraint in constraints:
                try:
                    session.run(constraint)
                    print(f"Constraint created: {constraint}")
                except Exception as e:
                    print(f"Error creating constraint: {constraint}\n{e}")

    def create_paper(self, tx, paper):
        """Insert a single paper into the database with a dynamic label."""

        query = f"""
        MERGE (p:Paper {{id: $id}})
        SET p.title = $title, p.year = $year, p.doi = $doi, 
            p.is_oa = $is_oa, p.oa_status = $oa_status, 
            p.ref_count = $ref_count, p.cited_by_count = $cited_by_count,
            p.type = $type, p.type_by4o = $type_by4o,
            p.measure_1 = $measure_1, p.measure_2 = $measure_2,
            p.category_1 = $category_1, p.category_2 = $category_2,
            p.cluster = $cluster,
            p.embedding = $embedding
        """
        tx.run(query, **paper)

    def create_year(self, tx, year):
        """Insert a single year node."""
        query = "MERGE (y:Year {year: $year})"
        tx.run(query, year=year)

    def create_measure(self, tx, measure):
        """Insert a single measure node."""
        query = "MERGE (m:Measure {name: $name})"
        tx.run(query, name=measure)

    def create_category(self, tx, category):
        """Insert a single category node."""
        query = "MERGE (c:Category {name: $name})"
        tx.run(query, name=category)
        
    def create_cluster(self, tx, cluster):
        """Insert a single cluster node."""
        query = "MERGE (cl:Cluster {id: $id})"
        tx.run(query, id=cluster)

    def create_relationships(self, tx, paper_rel):
        """Create multiple relationships for a single paper."""
        query = """
        MERGE (p:Paper {id: $id})
        
        FOREACH (y IN $year |
        MERGE (yearNode:Year {year: y})
        MERGE (p)-[:PUBLISHED_IN]->(yearNode)
        )

        FOREACH (cited_id IN $cited_papers |
            MERGE (cited_paper:Paper {id: cited_id})
            MERGE (cited_paper)-[:CITES]->(p)
        )

        FOREACH (measure IN $measures |
            MERGE (m:Measure {name: measure})
            MERGE (p)-[:USES]->(m)
        )

        FOREACH (category IN $categories |
            MERGE (c:Category {name: category})
            MERGE (p)-[:IS_CATEGORIZED_AS]->(c)
        )
        
        FOREACH (cluster IN $clusters |
            MERGE (cl:Cluster {id: cluster})
            MERGE (p)-[:BELONGS_TO]->(cl)
        )
        """
        tx.run(query, **paper_rel)


# Helper function to extract OpenAlex IDs from citations
def extract_openalex_ids(cited_by_str):
    if pd.isna(cited_by_str) or cited_by_str.strip() == "":
        return []
    return [url.strip() for url in cited_by_str.split(",") if url.strip()]


# Initialize Neo4j connection
neo4j_db = Neo4jDatabase(URI, USERNAME, PASSWORD)

# Create indexes before inserting data
neo4j_db.create_indexes()

# Insert Papers FIRST
print("Inserting Papers...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Papers"):
        paper = {
            "id": row["id"],
            "title": row.get("title", ""),
            "year": row.get("year", 0),
            "doi": row.get("doi", ""),
            "is_oa": row.get("is_oa", False),
            "oa_status": row.get("oa_status", ""),
            "ref_count": row.get("ref_count", 0),
            "cited_by_count": row.get("cited_by_count", 0),
            "type_by4o": row["sort_gpt4o_1"],  # Ensure safe label
            "type": row.get("type", ""),
            "category_1": row.get("category_1", ""),
            "category_2": row.get("category_2", ""),
            "measure_1": row.get("measure_1", ""),
            "measure_2": row.get("measure_2", ""),
            "cluster": row.get("measure_clusters", -1),
            "embedding": row.get("measure_embedding", None)
        }
        session.write_transaction(neo4j_db.create_paper, paper)

# Insert Year Nodes
print("Inserting Years...")
with neo4j_db.driver.session() as session:
    for year in tqdm(df["year"].dropna().unique(), desc="Years"):
        if year > 0:  # Remove invalid years
            session.write_transaction(neo4j_db.create_year, year)

# Insert Measure Nodes
print("Inserting Measures...")
with neo4j_db.driver.session() as session:
    for measure in tqdm(set(df["measure_1"].dropna().unique()) | set(df["measure_2"].dropna().unique()), desc="Measures"):
        if measure.strip():
            session.write_transaction(neo4j_db.create_measure, measure)


# Insert Category Nodes (handling NaN values properly)
print("Inserting Categories...")
with neo4j_db.driver.session() as session:
    categories = set(df["category_1"].dropna().astype(str).unique()) | set(df["category_2"].dropna().astype(str).unique())

    for category in tqdm(categories, desc="Categories"):
        if category and category.strip().lower() not in {"nan", "none"}:  # Ensure it's not NaN
            session.write_transaction(neo4j_db.create_category, category)

# Insert Cluster Nodes 
print("Inserting Clusters...")
with neo4j_db.driver.session() as session:
    for cluster in tqdm(df["measure_clusters"].dropna().unique(), desc="Clusters"):
        session.write_transaction(neo4j_db.create_cluster, cluster)


# Insert Relationships AFTER ensuring all nodes exist
print("Inserting Relationships...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Relationships"):
        paper_rel = {
            "id": row["id"],
            "year": row.get("year", 0),
            "cited_papers": extract_openalex_ids(row.get("cited_by", "")),
            "clusters": row.get("measure_clusters", -1),
            "embedding": row.get("measure_embedding", None),
            "measures": [m for m in [row.get("measure_1", ""), row.get("measure_2", "")] if pd.notna(m) and m.strip()],
            "categories": [c for c in [row.get("category_1", ""), row.get("category_2", "")] if pd.notna(c) and c.strip().lower() not in {"nan", "none"}]
        }

        session.write_transaction(neo4j_db.create_relationships, paper_rel)

# Close the database connection
neo4j_db.close()

print("Data successfully inserted into Neo4j!")
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.colors as pc

# Create unique nodes
nodes = list(pd.unique(cluster_citation[['citing_cluster', 'cited_cluster']].values.ravel()))
node_dict = {name: i for i, name in enumerate(nodes)}

# Create links
sources = [node_dict[c1] for c1 in cluster_citation['citing_cluster']]
targets = [node_dict[mc] for mc in cluster_citation['cited_cluster']]
values = list(cluster_citation['citation_count'])

# Select an appropriate color palette
color_palette = pc.qualitative.Set2
if len(sankey_counts['category'].unique()) > len(color_palette):
    color_palette = pc.qualitative.Dark24

# Assign colors to categories and clusters
unique_categories = cluster_citation['citing_cluster'].unique()
unique_clusters = cluster_citation['cited_cluster'].unique()

category_colors = {cat: color_palette[i % len(color_palette)] for i, cat in enumerate(unique_categories)}
cluster_colors = {cl: color_palette[(i + len(unique_categories)) % len(color_palette)] for i, cl in enumerate(unique_clusters)}

# Merge into one color dictionary
node_colors = {**category_colors, **cluster_colors}

# Function to convert rgb to rgba with opacity
def plotly_rgb_to_rgba(rgb_str, alpha=0.4):
    """Convert Plotly's 'rgb(r, g, b)' string to 'rgba(r, g, b, alpha)'."""
    rgb_values = rgb_str.replace("rgb(", "").replace(")", "").split(",")
    return f'rgba({rgb_values[0]},{rgb_values[1]},{rgb_values[2]},{alpha})'

# Assign colors to links based on the category source in sankey_long
link_colors = [plotly_rgb_to_rgba(node_colors[cat], alpha=0.4) for cat in cluster_citation['citing_cluster']]

# Construct the Sankey diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
        color=[node_colors[n] for n in nodes]  # Assign node colors
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=link_colors  # Lower opacity for links
    )
))

# Show the Sankey diagram
fig.show()
```

```python
# Function to process JSON and extract values
def parse_top5_measures(data, top_range=range(1, 6)):
    try:
        parsed = json.loads(data)
        result = {}
        for i in top_range:
            key = f"Top {i}"
            if key in parsed:
                entry = parsed[key]
                if isinstance(entry, dict):
                    measure = entry.get("Measure", entry.get("measure", "Unknown"))
                    freq = entry.get("Frequency", entry.get("frequency", "Unknown"))
                else:
                    # Extract measure and frequency from string
                    measure, freq = entry.rsplit(",", 1)
                    measure = measure.strip()
                    freq = int(freq.replace("Frequency", "").strip("() "))
                result[f"top{i}"] = measure
                result[f"freq_{i}"] = freq
            else:
                result[f"top{i}"] = "Unknown"
                result[f"freq_{i}"] = "Unknown"
        return result
    except Exception:
        return {
            "top1": "Error", "freq_1": "Error",
            "top2": "Error", "freq_2": "Error",
            "top3": "Error", "freq_3": "Error",
            "top4": "Error", "freq_4": "Error",
            "top5": "Error", "freq_5": "Error"
        }

# Create DataFrame
top5_period_df = pd.DataFrame(top5_period.items(), columns=['period', 'top5_measures'])

# Apply parsing function
parsed_data = top5_period_df['top5_measures'].apply(parse_top5_measures)
parsed_df = pd.DataFrame(parsed_data.tolist())

# Combine with original DataFrame
top5_period_df = pd.concat([top5_period_df.drop(columns=['top5_measures']), parsed_df], axis=1)
```

```python
def download_springer(df):
    for doi in df["doi"]:
        pdf_url = f"http://link.springer.com/{doi}.pdf"
        output_filename = f"{doi.replace('/', '_')}.pdf"  # Replace '/' to avoid path issues
        try:
            curl_command = f'curl -X GET -L "{pdf_url}" > "{output_filename}"'
            subprocess.run(curl_command, shell=True)
            print(f"Downloaded: {output_filename}")
        except Exception as e:
            print(f"Failed to download: {output_filename}")
            
```

```python
def doi_exists(doi):
    pdf_url = f"http://link.springer.com/{doi}.pdf"
    response = requests.head(pdf_url, allow_redirects=True)  # HEAD request
    if response.status_code == 200:
        print(f"Downloadable: {doi}")
    # return response.status_code == 200
```

```python
for doi in data['doi']:
    doi_exists(doi)
```

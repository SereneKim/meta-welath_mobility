````python
import re

def clean_abs(abs_text):
    abs_text = re.sub(r'[^\x00-\x7F]+', '', abs_text)  # Remove non-ASCII characters (like Korean)
    abs_text = re.sub(r'\bp\.\s*\d+\b', '', abs_text)  # Remove page numbers
    abs_text = re.sub(r'[A-Z][a-z]+(?:\s+(?:and|&)\s+[A-Z][a-z]+)*(?:,\s+[A-Z][a-z]+)*\s*\(\d{4}\)', '', abs_text)  # Remove parenthesis citation
    abs_text = re.sub(r'[A-Z][a-z]+(?:\s+(?:and|&)\s+[A-Z][a-z]+)*(?:,\s+[A-Z][a-z]+)*(?:,\s*\d{4})', '', abs_text)  # Remove text citation
    
    abs_text = re.sub(r'<[^>]+>', '', abs_text)  # Remove HTML tags
    abs_text = re.sub(r'\b(?:ABSTRACT|Abstract)\b', '', abs_text) # Remove the term 'ABSTRACT' or 'Abstract' (case-sensitive, preserving other variations)
    abs_text = re.sub(r'https?://\S+', '', abs_text)
    abs_text = re.sub(r'www\.\S+', '', abs_text)
    abs_text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b', '', abs_text)
    abs_text = re.sub('intergenerational elasticity', 'IGE', abs_text, flags=re.IGNORECASE)
    # Normalize white spaces
    abs_text = re.sub(r'(?<!\w)[.,]+(?!\w)', '', abs_text)  # Remove isolated periods and commas
    abs_text = re.sub(r'\s+', ' ', abs_text).strip()
    return abs_text

dic = {}
for i, row in data.iterrows():
    abs_text = row['abs'].strip()     
    dic[str(i)] = clean_abs(abs_text)
````

````python
df = pd.read_csv(f'{path}/data_abstracts/true_mobility_studies_617_forNeo4j.csv')

from neo4j import GraphDatabase
import pandas as pd
from tqdm import tqdm
import re

# Database connection details
URI = "bolt://localhost:7687"
USERNAME = "neo4j"
PASSWORD = "your_password"

class Neo4jDatabase:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def create_indexes(self):
        """Create indexes and constraints for better performance."""
        constraints = [
            "CREATE CONSTRAINT IF NOT EXISTS FOR (p:Paper) REQUIRE p.id IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (m:Measure) REQUIRE m.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (y:Year) REQUIRE y.year IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (c:Category) REQUIRE c.name IS UNIQUE;",
        ]

        with self.driver.session() as session:
            for constraint in constraints:
                try:
                    session.run(constraint)
                    print(f"Constraint created: {constraint}")
                except Exception as e:
                    print(f"Error creating constraint: {constraint}\n{e}")

    def create_paper(self, tx, paper):
        """Insert a single paper into the database with a dynamic label."""
        # label = paper["type"]  

        query = f"""
        MERGE (p:Paper {{id: $id}})
        SET p.title = $title, p.year = $year, p.doi = $doi, 
            p.is_oa = $is_oa, p.oa_status = $oa_status, 
            p.ref_count = $ref_count, p.cited_by_count = $cited_by_count,
            p.type = $type, p.type_by4o = $type_by4o,
            p.measure_1 = $measure_1, p.measure_2 = $measure_2,
            p.category_1 = $category_1, p.category_2 = $category_2,
            p.measure = $measure, p.data_cat = $data_cat, p.rq_cat = $rq_cat
        """
        tx.run(query, **paper)

    def create_year(self, tx, year):
        """Insert a single year node."""
        query = "MERGE (y:Year {year: $year})"
        tx.run(query, year=year)

    def create_measure(self, tx, measure):
        """Insert a single measure node."""
        query = "MERGE (m:Measure {name: $name})"
        tx.run(query, name=measure)

    def create_category(self, tx, category):
        """Insert a single category node."""
        query = "MERGE (c:Category {name: $name})"
        tx.run(query, name=category)

    def create_relationships(self, tx, paper_rel):
        """Create multiple relationships for a single paper."""
        query = """
        MERGE (p:Paper {id: $id})
        
        FOREACH (y IN $year |
        MERGE (yearNode:Year {year: y})
        MERGE (p)-[:PUBLISHED_IN]->(yearNode)
        )

        FOREACH (citing_id IN $citing_papers |
            MERGE (citing_paper:Paper {id: citing_id})
            MERGE (citing_paper)-[:CITES]->(p)
        )

        FOREACH (measure IN $measures |
            MERGE (m:Measure {name: measure})
            MERGE (p)-[:USES]->(m)
        )

        FOREACH (category IN $categories |
            MERGE (c:Category {name: category})
            MERGE (p)-[:IS_CATEGORIZED_AS]->(c)
        )
        """
        tx.run(query, **paper_rel)


# Helper function to extract OpenAlex IDs from citations
def extract_openalex_ids(cited_by_str):
    if pd.isna(cited_by_str) or cited_by_str.strip() == "":
        return []
    return [url.strip() for url in cited_by_str.split(",") if url.strip()]


# Initialize Neo4j connection
neo4j_db = Neo4jDatabase(URI, USERNAME, PASSWORD)

# Create indexes before inserting data
neo4j_db.create_indexes()

def safe_value(val, default=None):
    """
    Converts NaN to None (or a specified default).
    Trims strings and returns as-is for other data types.
    """
    if pd.isna(val):
        return default
    if isinstance(val, str):
        trimmed = val.strip()
        return trimmed if trimmed else default
    return val


# Insert Papers FIRST
print("Inserting Papers...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Papers"):
        # paper = {
        #     "id": row["id"],
        #     "title": row.get("title", ""),
        #     "year": row.get("year", 0),
        #     "doi": row.get("doi", ""),
        #     "is_oa": row.get("is_oa", False),
        #     "oa_status": row.get("oa_status", ""),
        #     "ref_count": row.get("ref_count", 0),
        #     "cited_by_count": row.get("cited_by_count", 0),
        #     "type_by4o": row["sort_gpt4o_1"],  
        #     "type": row.get("type", ""),
        #     "category_1": row.get("category_1", ""),
        #     "category_2": row.get("category_2", ""),
        #     "measure_1": row.get("measure_1", ""),
        #     "measure_2": row.get("measure_2", ""),
        #     "measure": row.get("measure", ""),
        #     "rq_cat": row.get("rq_cat", ""),
        #     "data_cat": row.get("data_cat", ""),
        # }
        paper = {
            "id": safe_value(row["id"]),
            "title": safe_value(row.get("title"), ""),
            "year": safe_value(row.get("year"), 0),
            "doi": safe_value(row.get("doi"), ""),
            "is_oa": row.get("is_oa", False),   # Assuming this is boolean already
            "oa_status": safe_value(row.get("oa_status"), ""),
            "ref_count": safe_value(row.get("ref_count"), 0),
            "cited_by_count": safe_value(row.get("cited_by_count"), 0),
            "type_by4o": safe_value(row.get("sort_gpt4o_1"), ""),
            "type": safe_value(row.get("type"), ""),
            "category_1": safe_value(row.get("category_1")),
            "category_2": safe_value(row.get("category_2")),
            "measure_1": safe_value(row.get("measure_1")),
            "measure_2": safe_value(row.get("measure_2")),
            "measure": safe_value(row.get("measure")),
            "rq_cat": safe_value(row.get("rq_cat")),
            "data_cat": safe_value(row.get("data_cat")),
        }
        session.write_transaction(neo4j_db.create_paper, paper)

# Insert Year Nodes
print("Inserting Years...")
with neo4j_db.driver.session() as session:
    for year in tqdm(df["year"].dropna().unique(), desc="Years"):
        if year > 0:  # Remove invalid years
            session.write_transaction(neo4j_db.create_year, year)

# Insert Measure Nodes
print("Inserting Measures...")
with neo4j_db.driver.session() as session:
    for measure in tqdm(set(df["measure_1"].dropna().unique()) | set(df["measure_2"].dropna().unique()), desc="Measures"):
        if measure.strip():
            session.write_transaction(neo4j_db.create_measure, measure)


# Insert Category Nodes (handling NaN values properly)
print("Inserting Categories...")
with neo4j_db.driver.session() as session:
    categories = set(df["category_1"].dropna().astype(str).unique()) | set(df["category_2"].dropna().astype(str).unique())

    for category in tqdm(categories, desc="Categories"):
        if category and category.strip().lower() not in {"nan", "none"}:  # Ensure it's not NaN
            session.write_transaction(neo4j_db.create_category, category)

# Insert Relationships AFTER ensuring all nodes exist
print("Inserting Relationships...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Relationships"):
        paper_rel = {
            "id": row["id"],
            "year": row.get("year", 0),
            "citing_papers": extract_openalex_ids(row.get("cited_by", "")),
            "measures": [
                safe_value(m) for m in [row.get("measure_1"), row.get("measure_2")]
                if safe_value(m)  # This filters out None and empty strings
            ],

            "categories": [
                safe_value(c) for c in [row.get("category_1"), row.get("category_2")]
                if safe_value(c) and safe_value(c).lower() not in {"nan", "none"}
            ]
            # "measures": [m for m in [row.get("measure_1", ""), row.get("measure_2", "")] if pd.notna(m) and m.strip()],
            # "categories": [c for c in [row.get("category_1", ""), row.get("category_2", "")] if pd.notna(c) and c.strip().lower() not in {"nan", "none"}]
        }

        session.write_transaction(neo4j_db.create_relationships, paper_rel)

# Close the database connection
neo4j_db.close()

print("Data successfully inserted into Neo4j!")
````

````python
import pandas as pd
from neo4j import GraphDatabase
from tqdm import tqdm
import ast

# Database connection details
URI = "bolt://localhost:7687"
USERNAME = "neo4j"
PASSWORD = "your_password"

df = pd.read_csv(f'{path}/data_abstracts/true_mobility_studies_617_forKGs.csv')

class Neo4jDatabase:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def create_indexes(self):
        constraints = [
            "CREATE CONSTRAINT IF NOT EXISTS FOR (p:Paper) REQUIRE p.id IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (m:Measure) REQUIRE m.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (y:Year) REQUIRE y.year IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (c:Category) REQUIRE c.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (t:Topic) REQUIRE t.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (f:Field) REQUIRE f.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (s:Subfield) REQUIRE s.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (a:Author) REQUIRE a.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (af:Affiliation) REQUIRE af.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (dc:DataCategory) REQUIRE dc.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (rq:ResearchQuestion) REQUIRE rq.name IS UNIQUE;",
            "CREATE CONSTRAINT IF NOT EXISTS FOR (oa:OpenAccess) REQUIRE oa.status IS UNIQUE;",
        ]

        with self.driver.session() as session:
            for constraint in constraints:
                try:
                    session.run(constraint)
                    print(f"Constraint created: {constraint}")
                except Exception as e:
                    print(f"Error creating constraint: {constraint}\n{e}")

    def create_node_by_label(self, tx, label, key, value):
        query = f"MERGE (n:{label} {{{key}: $value}})"
        tx.run(query, value=value)

    def create_paper(self, tx, paper):
        query = f"""
        MERGE (p:Paper {{id: $id}})
        SET p.title = $title, p.year = $year, p.doi = $doi, 
            p.ref_count = $ref_count, p.cited_by_count = $cited_by_count,
            p.type = $type, p.type_by4o = $type_by4o,
            p.measure = $measure, p.is_oa = $is_oa,
            p.data_type = $data_type, p.rq = $RQ
        """
        tx.run(query, **paper)

    def create_relationships(self, tx, paper_rel):
        query = """
        MERGE (p:Paper {id: $id})

        FOREACH (y IN $year |
            MERGE (yearNode:Year {year: y})
            MERGE (p)-[:PUBLISHED_IN]->(yearNode)
        )

        FOREACH (citing_id IN $citing_papers |
            MERGE (citing_paper:Paper {id: citing_id})
            MERGE (citing_paper)-[:CITES]->(p)
        )

        FOREACH (measure IN $measures |
            MERGE (m:Measure {name: measure})
            MERGE (p)-[:USES]->(m)
        )

        FOREACH (category IN $categories |
            MERGE (c:Category {name: category})
            MERGE (p)-[:IS_CATEGORIZED_AS]->(c)
        )

        
        FOREACH (pair IN $topic_score_pairs |
            MERGE (t:Topic {name: pair.topic})
            MERGE (p)-[r:HAS_TOPIC]->(t)
            SET r.score = pair.score
        )


        FOREACH (field IN $fields |
            MERGE (f:Field {name: field})
            MERGE (p)-[:HAS_FIELD]->(f)
        )

        FOREACH (subfield IN $subfields |
            MERGE (s:Subfield {name: subfield})
            MERGE (p)-[:HAS_SUBFIELD]->(s)
        )
        
        // Updated author-affiliation relationship creation
        FOREACH (pair IN $author_aff_pairs |
            MERGE (a:Author {name: pair.author})
            MERGE (p)-[:AUTHORED_BY]->(a)
            FOREACH (affiliation IN pair.affiliations |
                MERGE (af:Affiliation {name: affiliation})
                MERGE (a)-[:AFFILIATED_WITH]->(af)
            )
        )

        FOREACH (data_cat IN $data_cat |
            MERGE (dc:DataCategory {name: data_cat})
            MERGE (p)-[:HAS_DATA_CATEGORY]->(dc)
        )

        FOREACH (rq IN $rq_cat |
            MERGE (r:ResearchQuestion {name: rq})
            MERGE (p)-[:ADDRESSES]->(r)
        )

        FOREACH (oa IN $oa_status |
            MERGE (o:OpenAccess {status: oa})
            MERGE (p)-[:HAS_OPEN_ACCESS]->(o)
        )
        """
        tx.run(query, **paper_rel)

def safe_value(val, default=None):
    if pd.isna(val):
        return default
    if isinstance(val, str):
        trimmed = val.strip()
        return trimmed if trimmed else default
    return val

def extract_openalex_ids(cited_by_str):
    if pd.isna(cited_by_str) or cited_by_str.strip() == "":
        return []
    return [url.strip() for url in cited_by_str.split(",") if url.strip()]

def clean_list(raw_items):
    """Cleans and filters a list of values, removing None, NaN, empty, or 'none'."""
    return [
        str(item).strip()
        for item in raw_items
        if item is not None and str(item).strip().lower() not in {"", "nan", "none"}
    ]

def safe_eval_list(str_list):
    """Safely convert string representations of lists to actual lists."""
    if pd.isna(str_list) or not str_list:
        return []
    
    if isinstance(str_list, list):
        return str_list
        
    # Handle string representation of lists
    if isinstance(str_list, str):
        try:
            # Using ast.literal_eval is safer than eval()
            result = ast.literal_eval(str_list)
            if isinstance(result, list):
                return result
        except (SyntaxError, ValueError):
            # If parsing fails, try simpler approaches
            pass
    
    # Fall back to simple parsing if the above failed
    return [s.strip() for s in str(str_list).split(",") if s.strip()]

def process_author_affiliations(row):
    """Process author and affiliation data to create proper pairs."""
    author_data = row.get("author_raw_names")
    affiliation_data = row.get("author_raw_affiliations")
    
    # Convert string representations of lists to actual lists
    authors = safe_eval_list(author_data)
    
    # Handle the nested list structure of affiliations
    if pd.isna(affiliation_data) or not affiliation_data:
        affiliations = []
    else:
        try:
            # Try to parse the string representation of a list of lists
            raw_affiliations = ast.literal_eval(affiliation_data) if isinstance(affiliation_data, str) else affiliation_data
            
            # Ensure it's a list of lists
            if isinstance(raw_affiliations, list):
                # If it's already a list of lists, use it
                if all(isinstance(item, list) for item in raw_affiliations):
                    affiliations = raw_affiliations
                else:
                    # If it's a flat list, convert to list of lists
                    affiliations = [item if isinstance(item, list) else [item] for item in raw_affiliations]
            else:
                affiliations = []
        except (SyntaxError, ValueError):
            # Fallback for parsing errors
            print(f"Parsing error for affiliations: {affiliation_data}")
            affiliations = []
    
    # Create author-affiliation pairs
    author_aff_pairs = []
    
    # Case 1: We have authors
    if authors:
        for idx, author in enumerate(authors):
            # Get author's affiliations if available
            author_affiliations = []
            if idx < len(affiliations):
                # Make sure we handle both strings and lists
                affs = affiliations[idx]
                if isinstance(affs, list):
                    author_affiliations = clean_list(affs)
                else:
                    author_affiliations = [affs.strip()] if affs else []
            
            # Clean the author name (remove quotes or extra formatting)
            clean_author = author.strip("'\" ")
            
            # Add the pair
            author_aff_pairs.append({
                "author": clean_author,
                "affiliations": author_affiliations,
                "index": idx
            })
    
    return author_aff_pairs

def parse_string_list_field(value):
    """
    Parse fields that might contain lists or strings with values like:
    ['Regional Economics and Spatial Analysis'] -> 'Regional Economics and Spatial Analysis'
    Also removes duplicates within the same list.
    """
    if pd.isna(value) or not value:
        return []
    
    results = []
    
    # If it's already a list
    if isinstance(value, list):
        results = [item.strip("'\" ") for item in value if item]
    
    # If it's a string representation of a list
    elif isinstance(value, str):
        # Try to parse as literal Python list
        try:
            parsed = ast.literal_eval(value)
            if isinstance(parsed, list):
                # Extract the actual strings without the quotes
                results = [item.strip("'\" ") for item in parsed if item]
            else:
                # It's a single item - just return it
                results = [parsed.strip("'\" ")]
        except (SyntaxError, ValueError):
            # Not a Python list format, split by comma
            results = [item.strip() for item in value.split(",") if item.strip()]
    
    # Fall back to single item if we have no results yet
    else:
        results = [str(value)]
    
    # Remove duplicates while preserving order (using dict.fromkeys trick)
    return list(dict.fromkeys(results))

def process_topic_scores(row):
    topics = parse_string_list_field(row.get("topic_display_names", ""))
    scores = safe_eval_list(row.get("topic_scores", ""))

    topic_score_pairs = []
    if len(topics) == len(scores):
        for t, s in zip(topics, scores):
            try:
                topic_score_pairs.append({
                    "topic": t,
                    "score": float(s)
                })
            except ValueError:
                continue  # skip non-numeric scores
    else:
        print(f"[WARN] Topic/score length mismatch in row {row.get('id')}: {len(topics)} vs {len(scores)}")

    return topic_score_pairs


# Init DB
neo4j_db = Neo4jDatabase(URI, USERNAME, PASSWORD)
neo4j_db.create_indexes()

# Insert Paper Nodes
print("Inserting Papers...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Papers"):
        paper = {
            "id": safe_value(row["id"]),
            "title": safe_value(row.get("title"), ""),
            "doi": safe_value(row.get("doi"), ""),
            "ref_count": safe_value(row.get("ref_count"), 0),
            "cited_by_count": safe_value(row.get("cited_by_count"), 0),
            "type_by4o": safe_value(row.get("sort_gpt4o_1"), ""),
            "type": safe_value(row.get("type"), ""),
            "measure": safe_value(row.get("measure")), # it's the description of the measure
            "data_type": safe_value(row.get("data_type")),
            "RQ": safe_value(row.get("RQ")),
            "year": safe_value(row.get("year")),
            "is_oa": safe_value(row.get("is_oa", "")),
        }
        session.write_transaction(neo4j_db.create_paper, paper)

# Process topics, fields, and subfields correctly
all_topics = []
all_fields = []
all_subfields = []

# Extract and clean topics, fields, and subfields
print("Extracting topics, fields, and subfields...")
for _, row in tqdm(df.iterrows(), total=len(df), desc="Extracting metadata"):
    # Process topics
    if not pd.isna(row.get("topic_display_names")):
        topics = parse_string_list_field(row["topic_display_names"])
        all_topics.extend(topics)
    
    # Process fields
    if not pd.isna(row.get("field_display_names")):
        fields = parse_string_list_field(row["field_display_names"])
        all_fields.extend(fields)
    
    # Process subfields
    if not pd.isna(row.get("subfield_display_names")):
        subfields = parse_string_list_field(row["subfield_display_names"])
        all_subfields.extend(subfields)

# Remove duplicates and empty values
all_topics = [t for t in set(all_topics) if t and t.lower() not in {"nan", "none"}]
all_fields = [f for f in set(all_fields) if f and f.lower() not in {"nan", "none"}]
all_subfields = [s for s in set(all_subfields) if s and s.lower() not in {"nan", "none"}]

# Insert Node Types
print("Inserting Nodes...")
with neo4j_db.driver.session() as session:
    # Year nodes
    years = [y for y in df["year"].dropna().unique() if str(y).lower() not in {"nan", "none"}]
    for year in tqdm(years, desc="Year Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "Year", "year", year)
    
    # Measure nodes
    measures = set()
    for col in ["measure_1", "measure_2"]:
        measures.update([m for m in df[col].dropna().unique() if str(m).lower() not in {"nan", "none"}])
    for measure in tqdm(measures, desc="Measure Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "Measure", "name", measure)
    
    # Category nodes
    categories = set()
    for col in ["category_1", "category_2"]:
        categories.update([c for c in df[col].dropna().astype(str).unique() if c.lower() not in {"nan", "none"}])
    for category in tqdm(categories, desc="Category Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "Category", "name", category)
    
    # Topic nodes
    for topic in tqdm(all_topics, desc="Topic Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "Topic", "name", topic)
    
    # Field nodes
    for field in tqdm(all_fields, desc="Field Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "Field", "name", field)
    
    # Subfield nodes
    for subfield in tqdm(all_subfields, desc="Subfield Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "Subfield", "name", subfield)

    
    # Data Category nodes
    data_categories = [dc for dc in df["data_cat"].dropna().unique() if str(dc).lower() not in {"nan", "none"}]
    for dc in tqdm(data_categories, desc="Data Category Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "DataCategory", "name", dc)
    
    # Research Question nodes
    research_questions = [rq for rq in df["rq_cat"].dropna().unique() if str(rq).lower() not in {"nan", "none"}]
    for rq in tqdm(research_questions, desc="Research Question Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "ResearchQuestion", "name", rq)
    
    # Open Access Status nodes
    oa_statuses = [oa for oa in df["oa_status"].dropna().unique() if str(oa).lower() not in {"nan", "none"}]
    for oa in tqdm(oa_statuses, desc="OA Status Nodes"):
        session.write_transaction(neo4j_db.create_node_by_label, "OpenAccess", "status", oa)

# Insert Relationships
print("Inserting Relationships...")
with neo4j_db.driver.session() as session:
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Relationships"):
        # Process topics, fields, and subfields correctly
        topics = parse_string_list_field(row.get("topic_display_names", ""))
        fields = parse_string_list_field(row.get("field_display_names", ""))
        subfields = parse_string_list_field(row.get("subfield_display_names", ""))
        
        # Process author-affiliation pairs
        author_aff_pairs = process_author_affiliations(row) 
        topic_score_pairs = process_topic_scores(row)
        
        # Ensure unique values in all list attributes
        paper_rel = {
            "id": row["id"],
            "year": list(set([row["year"]])) if not pd.isna(row.get("year")) else [],
            "citing_papers": list(set(extract_openalex_ids(row.get("cited_by", "")))),
            "measures": list(set(clean_list([safe_value(row.get("measure_1")), safe_value(row.get("measure_2"))]))),
            "categories": list(set(clean_list([safe_value(row.get("category_1")), safe_value(row.get("category_2"))]))),
            "fields": list(set(fields)),
            "subfields": list(set(subfields)),
            "data_cat": list(set([row.get("data_cat")])) if not pd.isna(row.get("data_cat")) else [],
            "rq_cat": list(set([row.get("rq_cat")])) if not pd.isna(row.get("rq_cat")) else [],
            "oa_status": list(set([row.get("oa_status")])) if not pd.isna(row.get("oa_status")) else [],
            "author_aff_pairs": author_aff_pairs,  # This is already handled differently as it's a list of objects
            "topic_score_pairs": topic_score_pairs
        }
        session.write_transaction(neo4j_db.create_relationships, paper_rel)

# For debugging, print a few examples with duplicate checking
print("\n==== CHECKING FIELD PARSING AND DEDUPLICATION ====")
for _, row in df.head(3).iterrows():
    print(f"Paper ID: {row['id']}")
    
    # Topics
    original_topics = row.get('topic_display_names', 'None')
    parsed_topics = parse_string_list_field(original_topics)
    deduplicated_topics = list(set(parsed_topics))
    print(f"Original topics: {original_topics}")
    print(f"Parsed topics: {parsed_topics}")
    print(f"Deduplicated topics: {deduplicated_topics}")
    
    # Fields
    original_fields = row.get('field_display_names', 'None')
    parsed_fields = parse_string_list_field(original_fields)
    deduplicated_fields = list(set(parsed_fields))
    print(f"Original fields: {original_fields}")
    print(f"Parsed fields: {parsed_fields}")
    print(f"Deduplicated fields: {deduplicated_fields}")
    
    # Subfields
    original_subfields = row.get('subfield_display_names', 'None')
    parsed_subfields = parse_string_list_field(original_subfields)
    deduplicated_subfields = list(set(parsed_subfields))
    print(f"Original subfields: {original_subfields}")
    print(f"Parsed subfields: {parsed_subfields}")
    print(f"Deduplicated subfields: {deduplicated_subfields}")
    
    print("----------------------------")
print("==== END CHECKING ====\n")

neo4j_db.close()
print("Data successfully inserted into Neo4j!")
````

````python
from neo4j import GraphDatabase
import json
from datetime import datetime

class Neo4jEmbedder:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))
        self.timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        
        # Create output directory
        self.output_dir = f"graph_embeddings_{self.timestamp}"
        os.makedirs(self.output_dir, exist_ok=True)
        
    def close(self):
        self.driver.close()
        
    def check_gds_installation(self):
        """Verify GDS library is installed"""
        with self.driver.session() as session:
            try:
                result = session.run("CALL gds.list() YIELD name RETURN collect(name) AS algorithms")
                algorithms = result.single()["algorithms"]
                print(f"‚úì GDS library installed with {len(algorithms)} algorithms available")
                return True
            except Exception as e:
                print(f"‚úó Error accessing GDS library: {e}")
                print("Please install the Neo4j Graph Data Science library: https://neo4j.com/docs/graph-data-science/current/installation/")
                return False
                
                
    def project_full_graph(self):
        with self.driver.session() as session:
            # Drop existing graph if it exists
            session.run("CALL gds.graph.drop('full_graph', false) YIELD graphName")

            print("\nProjecting full graph with fixed schema...")

            projection_query = """
            CALL gds.graph.project(
                'full_graph',
                {
                    Paper: {properties: ['year', 'cited_by_count', 'ref_count']},
                    Author: {},
                    Topic: {},
                    Field: {},
                    Measure: {},
                    Category: {},
                    Subfield: {},
                    DataCategory: {},
                    ResearchQuestion: {},
                    Affiliation: {},
                    Year: {},
                    OpenAccess: {}
                },
                {
                    CITES: {type: 'CITES', orientation: 'NATURAL'},
                    AUTHORED_BY: {type: 'AUTHORED_BY', orientation: 'NATURAL'},
                    HAS_TOPIC: {type: 'HAS_TOPIC', orientation: 'NATURAL', properties: ['score']},
                    HAS_FIELD: {type: 'HAS_FIELD', orientation: 'NATURAL'},
                    USES: {type: 'USES', orientation: 'NATURAL'},
                    IS_CATEGORIZED_AS: {type: 'IS_CATEGORIZED_AS', orientation: 'NATURAL'},
                    HAS_SUBFIELD: {type: 'HAS_SUBFIELD', orientation: 'NATURAL'},
                    HAS_DATA_CATEGORY: {type: 'HAS_DATA_CATEGORY', orientation: 'NATURAL'},
                    ADDRESSES: {type: 'ADDRESSES', orientation: 'NATURAL'},
                    HAS_OPEN_ACCESS: {type: 'HAS_OPEN_ACCESS', orientation: 'NATURAL'},
                    AFFILIATED_WITH: {type: 'AFFILIATED_WITH', orientation: 'NATURAL'},
                    PUBLISHED_IN: {type: 'PUBLISHED_IN', orientation: 'NATURAL'}
                }
            )
            """

            try:
                result = session.run(projection_query)
                projection_stats = result.single()
                print(f"‚úì Graph projected with {projection_stats['nodeCount']} nodes and {projection_stats['relationshipCount']} relationships.")
                return True
            except Exception as e:
                print(f"‚úó Error projecting graph: {e}")
                return False

# Rest of the class implementation remains the same
    def generate_fastrp_embeddings(self, dimension=128, iterations=10):
        """Generate FastRP embeddings for the full graph"""
        with self.driver.session() as session:
            # Set up the parameters for FastRP
            print(f"\nGenerating FastRP embeddings (dimension={dimension}, iterations={iterations})...")
            
            embedding_query = """
            CALL gds.fastRP.write(
              'full_graph',
              {
                nodeLabels: ['Paper'],
                embeddingDimension: $dimension,
                randomSeed: 42,
                iterationWeights: $iterationWeights,
                writeProperty: 'embedding'
              }
            ) YIELD nodePropertiesWritten
            RETURN nodePropertiesWritten
            """
            
            # Create iteration weights array with proper length
            iteration_weights = [0.0] + [1.0] * (iterations - 1)
            
            try:
                result = session.run(
                    embedding_query, 
                    dimension=dimension,
                    iterationWeights=iteration_weights
                )
                stats = result.single()
                print(f"‚úì Generated embeddings for {stats['nodePropertiesWritten']} nodes")
                # Optionally, you can extract the embeddings by querying the graph directly
                query = """
                MATCH (p:Paper)
                WHERE p.embedding is not null
                RETURN p.id AS paper_id, p.embedding AS embedding
                """
                result = session.run(query)

                # Write the embeddings to a file
                with open(f"{self.output_dir}/fastRP_embeddings_fullgraph.json", "w") as f:
                    for record in result:
                        paper_id = record['paper_id']
                        embedding = record['embedding']
                        f.write(json.dumps({"paper_id": paper_id, "embedding": embedding}) + "\n")
                # Then later, I'll get the csv file of the 617 papers id to match.
                return True
            except Exception as e:
                print(f"‚úó Error generating embeddings: {e}")
                return False
            
    def detect_communities_LP(self):
        """Detect communities using the Label Propagation algorithm"""
        with self.driver.session() as session:
            print(f"\nRunning Label Propagation community detection...")

            community_query = """
            CALL gds.labelPropagation.write(
                'full_graph',
                {
                    writeProperty: 'community',
                    maxIterations: 50
                }
            ) YIELD communityCount
            RETURN communityCount
            """
            try:
                result = session.run(community_query)
                stats = result.single()
                print(f"‚úì Detected {stats['communityCount']} communities.")

                # Retrieve community assignment for each Paper node
                community_query = """
                MATCH (p:Paper)
                WHERE p.community IS NOT NULL
                RETURN p.id AS paper_id, p.community AS community
                """
                result = session.run(community_query)

                # Write community data to a file
                with open(f"{self.output_dir}/community_assignments.json", "w") as f:
                    for record in result:
                        paper_id = record['paper_id']
                        community = record['community']
                        f.write(json.dumps({"paper_id": paper_id, "community": community}) + "\n")
                return True
            except Exception as e:
                print(f"‚úó Error detecting communities: {e}")
                return False
 
    

# Run the analysis
def run_analysis(embedding_dim=128, iterations=10):
    embedder = Neo4jEmbedder(URI, USERNAME, PASSWORD)
    try:
        # Check GDS installation
        if not embedder.check_gds_installation():
            return
        # Project full graph
        if not embedder.project_full_graph():
            return
        # Generate FastRP embeddings
        # if not embedder.generate_fastrp_embeddings(dimension=embedding_dim, iterations=iterations):
        #     return
        # Detect communities using Label Propagation
        if not embedder.detect_communities_LP():
            return
    finally:
        embedder.close()

# Run the analysis with specified parameters
if __name__ == "__main__":
    run_analysis(embedding_dim=128, iterations=10)
````

````python
import os
import json
from datetime import datetime
from neo4j import GraphDatabase

class Neo4jEmbedder:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))
        self.timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        
        # Create output directory
        self.output_dir = f"graph_embeddings_{self.timestamp}"
        os.makedirs(self.output_dir, exist_ok=True)
        
    def close(self):
        self.driver.close()
        
    def check_gds_installation(self):
        """Verify GDS library is installed"""
        with self.driver.session() as session:
            try:
                result = session.run("CALL gds.list() YIELD name RETURN collect(name) AS algorithms")
                algorithms = result.single()["algorithms"]
                print(f"‚úì GDS library installed with {len(algorithms)} algorithms available")
                return True
            except Exception as e:
                print(f"‚úó Error accessing GDS library: {e}")
                print("Please install the Neo4j Graph Data Science library: https://neo4j.com/docs/graph-data-science/current/installation/")
                return False
                
    def project_full_graph(self):
        with self.driver.session() as session:
            # Drop existing graph if it exists
            session.run("CALL gds.graph.drop('full_graph', false) YIELD graphName")

            print("\nProjecting full graph with fixed schema...")

            projection_query = """
            CALL gds.graph.project(
                'full_graph',
                ['Paper', 'Author', 'Topic', 'Measure', 'Category', 'DataCategory', 'ResearchQuestion'],
                ['CITES', 'AUTHORED_BY', 'HAS_TOPIC', 'USES', 'IS_CATEGORIZED_AS', 'HAS_DATA_CATEGORY', 'ADDRESSES']
            )
            """

            try:
                result = session.run(projection_query)
                projection_stats = result.single()
                print(f"‚úì Graph projected with {projection_stats['nodeCount']} nodes and {projection_stats['relationshipCount']} relationships.")
                return True
            except Exception as e:
                print(f"‚úó Error projecting graph: {e}")
                return False

    def generate_fastrp_embeddings(self, dimension=128, iterations=10):
        """Generate FastRP embeddings for the full graph"""
        with self.driver.session() as session:
            # Set up the parameters for FastRP
            print(f"\nGenerating FastRP embeddings (dimension={dimension}, iterations={iterations})...")
            
            embedding_query = """
            CALL gds.fastRP.write(
              'full_graph',
              {
                nodeLabels: ['Paper'],
                embeddingDimension: $dimension,
                randomSeed: 42,
                iterationWeights: $iterationWeights,
                writeProperty: 'embedding'
              }
            ) YIELD nodePropertiesWritten
            RETURN nodePropertiesWritten
            """
            
            # Create iteration weights array with proper length
            iteration_weights = [0.0] + [1.0] * (iterations - 1)
            
            try:
                result = session.run(
                    embedding_query, 
                    dimension=dimension,
                    iterationWeights=iteration_weights
                )
                stats = result.single()
                print(f"‚úì Generated embeddings for {stats['nodePropertiesWritten']} nodes")
                
                # Query to retrieve the embeddings
                query = """
                MATCH (p:Paper)
                WHERE p.embedding IS NOT NULL
                RETURN p.id AS paper_id, p.embedding AS embedding
                """
                result = session.run(query)

                # Write the embeddings to a file
                with open(f"{self.output_dir}/fastRP_embeddings_fullgraph.json", "w") as f:
                    for record in result:
                        paper_id = record['paper_id']
                        embedding = record['embedding']
                        f.write(json.dumps({"paper_id": paper_id, "embedding": embedding}) + "\n")
                
                # Optionally, you can get the csv file of the 617 papers' ids to match.
                return True
            except Exception as e:
                print(f"‚úó Error generating embeddings: {e}")
                return False
    
    def detect_communities(self):
        """Detect communities using the Louvain algorithm"""
        with self.driver.session() as session:
            # Community detection using Louvain method
            print(f"\nRunning Louvain community detection...")

            community_query = """
            CALL gds.louvain.write(
                'full_graph',
                {
                    writeProperty: 'multiCommunity'
                }
            ) YIELD communityCount, modularity
            RETURN communityCount, modularity
            """
            try:
                result = session.run(community_query)
                stats = result.single()
                print(f"‚úì Detected {stats['communityCount']} communities with modularity {stats['modularity']}.")
                
                # Optionally, retrieve the community assignment for each node (Paper)
                community_query = """
                MATCH (p:Paper)
                WHERE p.multiCommunity IS NOT NULL
                RETURN p.id AS paper_id, p.multiCommunity AS community
                """
                result = session.run(community_query)

                # Write community data to a file
                with open(f"{self.output_dir}/community_assignments_{self.timestamp}.json", "w") as f:
                    for record in result:
                        paper_id = record['paper_id']
                        community = record['community']
                        f.write(json.dumps({"paper_id": paper_id, "community": community}) + "\n")
                
                return True
            except Exception as e:
                print(f"‚úó Error detecting communities: {e}")
                return False

# Run the analysis
def run_analysis(embedding_dim=128, iterations=10):
    embedder = Neo4jEmbedder("bolt://localhost:7687", "neo4j", "your_password")  # Update with your credentials
    try:
        # Check GDS installation
        if not embedder.check_gds_installation():
            return
        # Project full graph
        if not embedder.project_full_graph():
            return
        # Generate FastRP embeddings
        # if not embedder.generate_fastrp_embeddings(dimension=embedding_dim, iterations=iterations):
            # return
        if not embedder.detect_communities():
            return
    finally:
        embedder.close()

# Run the analysis with specified parameters
if __name__ == "__main__":
    run_analysis(embedding_dim=128, iterations=10)
````

## Implement it to the graph in Neo4j

````python
import pandas as pd
from neo4j import GraphDatabase

class Neo4jClusterUpdater:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))
        
    def close(self):
        self.driver.close()

    def update_clusters(self, csv_path):
        """Update cluster labels for Paper nodes in Neo4j"""
        # Read the CSV file with cluster labels
        df = pd.read_csv(csv_path)
        
        with self.driver.session() as session:
            # Iterate over each row in the DataFrame and update the corresponding Paper node
            for _, row in df.iterrows():
                paper_id = row['id']
                cluster = row['cluster']  # Cluster number
                
                # Update the cluster property on the Paper node based on the ID
                update_query = """
                MATCH (p:Paper {id: $id})
                SET p.cluster_simpler = $cluster
                """
                session.run(update_query, id=paper_id, cluster=cluster)
            
            print(f"Updated cluster labels for {len(df)} papers.")
        

# Example usage
def run_analysis():
    uri = "bolt://localhost:7687"
    user = "neo4j"
    password = "your_password" 
    # csv_path = f"{path}/graph_embeddings_20250507_134315/fastRP_embeddings_617_clusters.csv"
    # csv_path = f"{path}/graph_embeddings_20250507_220826/fastRP_embeddings_617_clusters.csv"
    csv_path = f"{path}/graph_embeddings_20250507_220826/fastRP_embeddings_617_clusters.csv"
    
    updater = Neo4jClusterUpdater(uri, user, password)
    try:
        updater.update_clusters(csv_path)
    finally:
        updater.close()

# Run the analysis with specified parameters
if __name__ == "__main__":
    run_analysis()
````
